# Linear Regression model Training 

### Diabetes Small datasets->


#### importing requred libraries :- [Matplotlib for data visualisation ,  Numpy for mathematical Computation , # Scikit Learn libraries --> from sklearn import datastes, linear_model , model_selection]

```python
import matplotlib.pyplot as plt
import numpy as np
from  sklearn  import datasets , linear_model , model_selection
```

```python
X,y  = datasets.load_diabetes(return_X_y=True)
print(X.shape) # returns shape of the datasets
print(X[0]) # returns 1st row values 
```

```python
X = X[:, 2]  # this line say take one future in column section (X)
X = X.reshape(-1,1) # reshaping from 1D array  to 2D array
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test , color='black')
plt.plot(X_test , y_pred , color='blue', label='Prediction')
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes progression Against BMI')
plt.legend()
plt.show()
```
